```python
#1) Write a function to convert temperature from Celsius to Fahrenheit and vice versa.
def convert_temperature(value, scale):
    if scale == "C":
        return (value * 9 / 5) + 32
    elif scale == "F":
        return (value - 32) * 5 / 9
    else:
        return "Invalid scale"

temp = float(input("Enter temperature value: "))
scale = input("Enter scale (C for Celsius, F for Fahrenheit): ").upper()

result = convert_temperature(temp, scale)
print("Converted temperature:", result)
```

```python
#2) Write a function that accepts a string and returns the count of vowels and consonants.
def count_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0

    for ch in s:
        if ch.isalpha():
            if ch in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    return vowel_count, consonant_count

text = input("Enter a string: ")
vowels, consonants = count_vowels_consonants(text)
print("Vowels:", vowels)
print("Consonants:", consonants)
```

```python
#3) Write a function that accepts variable number of arguments and returns their average.
def average(*numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)

nums = list(map(int, input("Enter numbers separated by space: ").split()))
print("Average:", average(*nums))
```

```python
#4) Write a lambda function to find the maximum of three numbers.
maximum = lambda a, b, c: a if (a >= b and a >= c) else (b if b >= c else c)

x = float(input("Enter first number: "))
y = float(input("Enter second number: "))
z = float(input("Enter third number: "))

print("Maximum number is:", maximum(x, y, z))
```
