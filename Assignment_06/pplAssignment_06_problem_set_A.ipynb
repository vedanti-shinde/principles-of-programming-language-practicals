```python
#1) Write a function to find the maximum of three numbers.
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))

print("Maximum number is:", find_max(x, y, z))
```

```python
#2) Write a function to check if a number is prime or not.
def is_prime(n):
    if n <= 1:                  # numbers <= 1 are not prime
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:          # divisible by a number other than 1 and itself
            return False
    return True

num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
#3) Write a function to calculate the factorial of a number (both iterative and recursive).
def factorial_iterative(n):         # Iterative factorial

    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def factorial_recursive(n):         # Recursive factorial
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)

num = int(input("Enter a number: "))

print("Factorial (Iterative):", factorial_iterative(num))
print("Factorial (Recursive):", factorial_recursive(num))
```

```python
#4) Write a function that accepts a list and returns a new list with unique elements.
def unique_elements(lst):
    return list(set(lst))  # convert list to set to remove duplicates and back to list

numbers = list(map(int, input("Enter elements separated by space: ").split()))
print("Unique elements:", unique_elements(numbers))
```

```python
#5) Write a function to calculate the sum of all even numbers in a list.
def sum_even_numbers(lst):
    total = 0
    for num in lst:
        if num % 2 == 0:
            total += num
    return total

numbers = list(map(int, input("Enter numbers separated by space: ").split()))
print("Sum of even numbers:", sum_even_numbers(numbers))
```
