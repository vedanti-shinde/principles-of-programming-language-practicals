```python
#1) Write a recursive function to calculate the sum of digits of a number.
def sum_of_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_of_digits(n // 10)

num = int(input("Enter a number: "))
print("Sum of digits of the number:", sum_of_digits(num))
```

```python
#2) Write a function that accepts a list of numbers and returns a dictionary with count of even and odd numbers.
def count_even_odd(numbers):
    result = {"even": 0, "odd": 0}

    for n in numbers:
        if n % 2 == 0:
            result["even"] += 1
        else:
            result["odd"] += 1

    return result

lst = list(map(int, input("Enter numbers separated by space: ").split()))
print(count_even_odd(lst))
```
