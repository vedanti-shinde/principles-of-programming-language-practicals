```python
#1) Write a program to find the longest word in a sentence
sentence = input("Enter a sentence: ")
words = sentence.split()
longest_word = ""

for word in words:
    if len(word) > len(longest_word):
        longest_word = word

print("Longest word is:", longest_word)
print("Length of longest word is:", len(longest_word))
```

```python
#2) Write a program to check if a string contains only digits, only alphabets, or both.

s = input("Enter a string: ")

if s.isdigit():
    print("String contains only digits")
elif s.isalpha():
    print("String contains only alphabets")
else:
    print("String contains both digits and alphabets")
```
