```python
#1) Create a class hierarchy: Animal -> Mammal -> Dog. Demonstrate multilevel inheritance
class Animal:                          # base class
    def speak(self):
        print("Animal makes a sound")

class Mammal(Animal):                 # derived from Animal
    def walk(self):
        print("Mammal can walk")

class Dog(Mammal):                    # derived from Mammal
    def bark(self):
        print("Dog barks")

d = Dog()                             # create Dog object

d.speak()                             # from Animal
d.walk()                              # from Mammal
d.bark()                              # from Dog
```

```python
#2) Create a class Employee with class method to keep count of total employees.
class Employee:
    total_employees = 0               # class variable

    def __init__(self, name):
        self.name = name
        Employee.total_employees += 1

    @classmethod
    def get_total_employees(cls):      # class method
        return cls.total_employees

e1 = Employee("Alice")
e2 = Employee("Bob")
e3 = Employee("Charlie")

print("Total Employees:", Employee.get_total_employees())
```

```python
#3) Create a class Book with special methods __str__, __len__, and __eq__.
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return f"Book Title: {self.title}, Pages: {self.pages}"

    def __len__(self):
        return self.pages

    def __eq__(self, other):
        return self.title == other.title and self.pages == other.pages

b1 = Book("Python Basics", 200)
b2 = Book("Python Basics", 200)

print(b1)
print("Number of pages:", len(b1))
print("Books are equal:", b1 == b2)
```

```python
#4) Create a class Temperature with property decorators for Celsius and Fahrenheit conversion.
class Temperature:
    def __init__(self, celsius=0):          # initialize with Celsius
        self._celsius = celsius

    @property
    def celsius(self):                      # getter for Celsius
        return self._celsius

    @celsius.setter
    def celsius(self, value):               # setter for Celsius
        self._celsius = value

    @property
    def fahrenheit(self):                   # getter for Fahrenheit
        return (self._celsius * 9/5) + 32

    @fahrenheit.setter
    def fahrenheit(self, value):            # setter for Fahrenheit
        self._celsius = (value - 32) * 5/9


temp = Temperature(25)
print("Celsius:", temp.celsius)
print("Fahrenheit:", temp.fahrenheit)

temp.fahrenheit = 98.6
print("Celsius after setting Fahrenheit:", temp.celsius)
```
