```python
'''1) Create an abstract class Shape with abstract method area(). Create subclasses Rectangle,
Circle, and Triangle that implement the area method.'''
from abc import ABC, abstractmethod

class Shape(ABC):           # abstract class Shape
    @abstractmethod
    def area(self):        # abstract method area
        pass

class Rectangle(Shape):           # Rectangle inherits Shape
    def __init__(self, length, breadth):          # constructor
        self.length = length
        self.breadth = breadth

    def area(self):            # area of rectangle
        return self.length * self.breadth

class Circle(Shape):            # Circle inherits Shape
    def __init__(self, radius):      # constructor
        self.radius = radius

    def area(self):            # area of circle
        return 3.14 * self.radius * self.radius

class Triangle(Shape):           # Triangle inherits Shape
    def __init__(self, base, height):      # constructor
        self.base = base
        self.height = height

    def area(self):              # area of triangle
        return 0.5 * self.base * self.height

r = Rectangle(5, 4)         # Rectangle object
c = Circle(3)                # Circle object
t = Triangle(6, 2)               # Triangle object

print("Rectangle Area:", r.area())   # print rectangle area
print("Circle Area:", c.area())      # print circle area
print("Triangle Area:", t.area())    # print triangle area
```

```python
'''2) Create a class Library that manages a collection of books. Implement methods to add
books, remove books, search books, and display all books.
'''
class Library:
    def __init__(self):           # constructor
        self.books = []           # list to store books

    def add_book(self, book):           # add a book
        self.books.append(book)
        print("Book added")

    def remove_book(self, book):          # remove a book
        if book in self.books:
            self.books.remove(book)
            print("Book removed")
        else:
            print("Book not found")

    def search_book(self, book):        # search for a book
        if book in self.books:
            print("Book found")
        else:
            print("Book not found")

    def display_books(self):          # display all books
        if len(self.books) == 0:
            print("No books in library")
        else:
            for book in self.books:
                print(book)

lib = Library()             # create Library object

lib.add_book("Python Basics")
lib.add_book("Data Science")
lib.add_book("Artificial Intelligence")

lib.display_books()

lib.search_book("Data Science")
lib.remove_book("Python Basics")

lib.display_books()
```
