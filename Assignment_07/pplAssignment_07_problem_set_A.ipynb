```python
#1) Create a class called Rectangle with attributes length and width. Add methods to calculate area and perimeter.
class Rectangle:
    def __init__(self, length, width):        # constructor
        self.length = length
        self.width = width

    def area(self):                           # calculate area
        return self.length * self.width

    def perimeter(self):                      # calculate perimeter
        return 2 * (self.length + self.width)


r = Rectangle(5, 3)                          # create Rectangle object

print("Area:", r.area())                      # print area
print("Perimeter:", r.perimeter())            # print perimeter
```

```python
#2) Create a class BankAccount with methods to deposit, withdraw, and display balance.
class BankAccount:
    def __init__(self, owner, balance=0):      # constructor with optional balance
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):                # deposit money
        if amount > 0:
            self.balance += amount
            print(f"{amount} deposited. New balance: {self.balance}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):               # withdraw money
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"{amount} withdrawn. New balance: {self.balance}")
            else:
                print("Insufficient balance")
        else:
            print("Invalid withdrawal amount")

    def display_balance(self):                # show balance
        print(f"Account owner: {self.owner}, Balance: {self.balance}")


acc = BankAccount("Alice", 1000)             # create BankAccount object

acc.deposit(500)
acc.withdraw(200)
acc.display_balance()
```

```python
'''3) Create a parent class Vehicle and child classes Car and Bike that inherit from it. Override
a method in child classes.'''
class Vehicle:                          # parent class
    def start(self):
        print("Vehicle is starting")

class Car(Vehicle):                     # Car inherits Vehicle
    def start(self):                     # overriding start method
        print("Car engine starts with a key")

class Bike(Vehicle):                    # Bike inherits Vehicle
    def start(self):                     # overriding start method
        print("Bike engine starts with a button")

v = Vehicle()
c = Car()
b = Bike()

v.start()                               # Vehicle method
c.start()                               # Car overridden method
b.start()                               # Bike overridden method
```

```python
#4) Create a class Student with private attributes and public methods to access them.
class Student:
    def __init__(self, name, age):          # constructor
        self.__name = name                  # private attribute
        self.__age = age                    # private attribute

    def get_name(self):                     # public method to access name
        return self.__name

    def set_name(self, name):               # public method to modify name
        self.__name = name

    def get_age(self):                      # public method to access age
        return self.__age

    def set_age(self, age):                 # public method to modify age
        self.__age = age


s = Student("Alice", 17)

print("Name:", s.get_name())               # access private attribute
print("Age:", s.get_age())

s.set_name("Bob")                          # modify private attribute
s.set_age(18)

print("Updated Name:", s.get_name())
print("Updated Age:", s.get_age())
```
