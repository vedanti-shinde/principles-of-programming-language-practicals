```python
#1) Write a program which accept an integer value 'n' and display all prime numbers till 'n'.
n = int(input("Enter a number: "))

print("Prime numbers up to", n, "are:")

for i in range(2, n+1):  # start from 2, because 1 is not prime
    is_prime = True
    for j in range(2, i):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        print(i, end=" ")
```

```python
#2) Write a program that accept two integer values and if both are equal then prints "SAME
#identity" otherwise prints "DIFFERENT identity".
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a == b:
    print("SAME identity")
else:
    print("DIFFERENT identity")
```

```python
#3) Write a program to display following pattern:
#1 2 3 4
#1 2 3
#1 2
#1
# Program to display the given pattern

rows = 4

for i in range(rows, 0, -1):  # start from 4 down to 1
    for j in range(1, i+1):
        print(j, end=" ")
    print()  # move to next line
```

```python
#4) Write a program to reverse a given number.
# Program to reverse a given number

n = int(input("Enter a number: "))
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print("Reversed number is:", rev)
```
