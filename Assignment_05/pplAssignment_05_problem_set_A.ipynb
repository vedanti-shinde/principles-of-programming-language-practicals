```python
#1) Write a program to create a dictionary with roll number as key and name as value for 5 students. Display all key-value pairs.
students = {}    # Create an empty dictionary

for i in range(5):      # Loop to input roll number and name for 5 students
    roll_no = input(f"Enter roll number of student {i+1}: ")
    name = input(f"Enter name of student {i+1}: ")
    students[roll_no] = name         # Add key-value pair to dictionary

print("Dictionary of students:")         # Display all key-value pairs

for roll, name in students.items():
    print(f"Roll Number: {roll}, Name: {name}")
```

```python
'''2) Write a program to count the frequency of each word in a given sentence using a
dictionary.
Example: "hello world hello" should give {'hello': 2, 'world': 1}'''
sentence = input("Enter a sentence: ")    # Input sentence

words = sentence.split()      # Split sentence into words

word_count = {}      # Create an empty dictionary

for word in words:      # Count frequency of each word
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("Word frequency:", word_count)
```

```python
#3) Write a program to merge two dictionaries entered by the user.
dict1 = {}    # First dictionary

n1 = int(input("Enter number of elements in first dictionary: "))

for i in range(n1):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict1[key] = value

dict2 = {}   # Second dictionary

n2 = int(input("\nEnter number of elements in second dictionary: "))

for i in range(n2):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict2[key] = value

# Merge dictionaries
merged_dict = dict1.copy()   # copy first dictionary
merged_dict.update(dict2)    # add second dictionary

print("\nMerged Dictionary:")
print(merged_dict)
```

```python
#4) Write a program to find the key with the maximum value in a dictionary.
my_dict = {}      # Create an empty dictionary


n = int(input("Enter number of elements in the dictionary: "))      # Input number of elements


for i in range(n):    # Input key-value pairs

    key = input("Enter key: ")
    value = int(input("Enter value: "))  # taking value as integer
    my_dict[key] = value

max_key = max(my_dict, key=my_dict.get)   # Find key with max value


print("Dictionary:", my_dict)
print("Key with maximum value:", max_key)
print("Maximum value:", my_dict[max_key])
```

```python
#5) Write a program to check if a given key exists in a dictionary.
my_dict = {}   # Create an empty dictionary

n = int(input("Enter number of elements in the dictionary: "))    # Input number of elements

for i in range(n):     # Input key-value pairs
    key = input("Enter key: ")
    value = input("Enter value: ")
    my_dict[key] = value

check_key = input("Enter the key to check: ")     # Input key to check 

if check_key in my_dict:      # Check if key exists
    print("The key exists in the dictionary.")
else:
    print("The key does NOT exist in the dictionary.")
```
