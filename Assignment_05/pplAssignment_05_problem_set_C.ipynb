```python
#1) Write a program to find all keys with the same value in a dictionary.
my_dict = {}

n = int(input("Enter number of elements: "))

for i in range(n):
    key = input("Enter key: ")
    value = input("Enter value: ")
    my_dict[key] = value

search_value = input("Enter value to search: ")

same_value_keys = []

for key in my_dict:
    if my_dict[key] == search_value:
        same_value_keys.append(key)

print("Keys with the same value:", same_value_keys)
```

```python
#2) Write a program to create a dictionary where keys are numbers from 1 to n and values are their cubes.
n = int(input("Enter a number: "))

my_dict = {}

for i in range(1, n + 1):
    my_dict[i] = i ** 3

print(my_dict)
```
