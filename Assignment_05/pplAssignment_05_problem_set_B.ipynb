```python
#1) Write a program to remove a key from a dictionary and handle the case when key doesn't exist
# Program to remove a key from a dictionary

my_dict = {}

n = int(input("Enter number of elements in the dictionary: "))        # Input number of elements

for i in range(n):    # Input key-value pairs
    key = input("Enter key: ")
    value = input("Enter value: ")
    my_dict[key] = value

# Display dictionary before deletion
print("\nDictionary before deletion:", my_dict)

# Input key to remove
remove_key = input("Enter the key to remove: ")

# Remove key safely
if remove_key in my_dict:
    del my_dict[remove_key]
    print("Key removed successfully.")
else:
    print("Key does not exist in the dictionary.")

# Display dictionary after deletion
print("Dictionary after deletion:", my_dict)
```

```python
#2) Write a program to sort a dictionary by its values in descending order.
my_dict = {}
n = int(input("Enter number of elements in the dictionary: "))
for i in range(n):
    key = input("Enter key: ")
    value = int(input("Enter value: ")) 
    my_dict[key] = value
sorted_dict = dict(sorted(my_dict.items(), key=lambda item: item[1], reverse=True))

print("\nOriginal Dictionary:", my_dict)
print("Dictionary sorted by values (descending):", sorted_dict)
```

```python
### 3) Write a program to create a dictionary from two lists - one containing keys and another containing values.
keys_input = input("Enter keys separated by space: ")
keys = keys_input.split()
print("The list containing keys is: ", keys)
values_input = input("Enter values separated by space: ")
values = values_input.split()
print("The list containing values is: ", values)

my_dict = dict(zip(keys, values))

print("Created Dictionary:", my_dict)
```
