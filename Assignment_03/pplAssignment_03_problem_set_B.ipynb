```python
#1) Write a program to find the second largest element in a tuple.
t = (10, 5, 20, 8, 15)

largest = t[0]
second_largest = t[0]

for num in t:
    if num > largest:
        second_largest = largest
        largest = num
    elif num != largest and num > second_largest:
        second_largest = num

print("Second largest element:", second_largest)
```

```python
'''2) Write a program to sort a tuple of tuples based on the second element.
Example: ((1, 5), (3, 2), (2, 8)) should become ((3, 2), (1, 5), (2, 8))'''
t = ((1, 5), (3, 2), (2, 8))
lst = list(t)

for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i][1] > lst[j][1]:
            lst[i], lst[j] = lst[j], lst[i]

result = tuple(lst)

print("Sorted tuple:", result)
```

```python
#3) Write a program to check if all elements in a tuple are unique.
t = (1, 2, 3, 4, 5)
unique = True

for i in range(len(t)):
    for j in range(i + 1, len(t)):
        if t[i] == t[j]:
            unique = False
            break

if unique:
    print("All elements are unique")
else:
    print("Tuple contains duplicate elements")
```

```python
#4) Write a program to swap the first and last element of a tuple.
t = (10, 20, 30, 40, 50)

t = (t[-1],) + t[1:-1] + (t[0],)

print("Tuple after swapping:", t)
```
