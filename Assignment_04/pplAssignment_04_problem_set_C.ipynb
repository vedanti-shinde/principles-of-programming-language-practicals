```python
'''1) Write a program that takes two sentences as input and finds:
● Common words in both sentences
● Words unique to the first sentence
● Words unique to the second sentence'''

s1 = input("Enter first sentence: ")
s2 = input("Enter second sentence: ")

words1 = s1.split()
words2 = s2.split()

# Common words
common = []
for w in words1:
    if w in words2 and w not in common:
        common.append(w)

# Words unique to first sentence
unique1 = []
for w in words1:
    if w not in words2 and w not in unique1:
        unique1.append(w)

# Words unique to second sentence
unique2 = []
for w in words2:
    if w not in words1 and w not in unique2:
        unique2.append(w)

print("Common words:", common)
print("Words unique to first sentence:", unique1)
print("Words unique to second sentence:", unique2)
```

```python
'''2) Write a program to perform all set operations (union, intersection, difference, symmetric
difference) on three sets.'''
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
C = {4, 6, 7, 8}

# Union
union_set = A | B | C

# Intersection
intersection_set = A & B & C

# Difference
diff_A_B = A - B
diff_A_C = A - C

# Symmetric Difference
sym_diff_AB = A ^ B
sym_diff_BC = B ^ C
sym_diff_AC = A ^ C

print("Union of A, B and C:", union_set)
print("Intersection of A, B and C:", intersection_set)
print("Difference of A and B:", diff_A_B)
print("Difference of A and C:", diff_A_C)
print("Symmetric Difference of A and B:", sym_diff_AB)
print("Symmetric Difference of B and C:", sym_diff_BC)
print("Symmetric Difference of A and C:", sym_diff_AC)
```
