```python
#1) Write a program to find symmetric difference between two sets.
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

sym_diff = A ^ B

print("Symmetric Difference is:", sym_diff)
```

```python
#2) Write a program to add multiple elements to a set at once from user input
s = set()

elements = input("Enter elements separated by space: ").split()

for item in elements:
    s.add(item)

print("Updated set:", s)
```

```python
#3) Write a program to find the length of a set and check if a specific element exists in it.
s = {10, 20, 30, 40, 50}

print("Length of set:", len(s))

element = int(input("Enter element to check: "))

if element in s:
    print("Element exists in the set")
else:
    print("Element does not exist in the set")
```
