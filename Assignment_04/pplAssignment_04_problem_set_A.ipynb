```python
#1) Write a program to find the union, intersection, and difference of two sets entered by the user.
# Input first set
set1_input = input("Enter elements of first set separated by space: ")
set1 = set(set1_input.split())  # Convert input to a set

# Input second set
set2_input = input("Enter elements of second set separated by space: ")
set2 = set(set2_input.split())  # Convert input to a set

# Union
union_set = set1.union(set2)
print("Union of sets:", union_set)

# Intersection
intersection_set = set1.intersection(set2)
print("Intersection of sets:", intersection_set)

# Difference (set1 - set2)
difference_set = set1.difference(set2)
print("Difference of first set from second set:", difference_set)
```

```python
#2) Write a program to check if a set is a subset of another set.
# Input first set
set1_input = input("Enter elements of the first set separated by space: ")
set1 = set(set1_input.split())

# Input second set
set2_input = input("Enter elements of the second set separated by space: ")
set2 = set(set2_input.split())

# Check if set1 is a subset of set2
if set1.issubset(set2):
    print("The first set is a subset of the second set.")
else:
    print("The first set is NOT a subset of the second set.")
```

```python
#3) Write a program to remove duplicates from a list using sets.
# Input list from user
list_input = input("Enter elements of the list separated by space: ")
my_list = list_input.split()  # Convert input to a list

# Remove duplicates by converting list to set
unique_set = set(my_list)

# Convert set back to list (optional, to keep it as a list)
unique_list = list(unique_set)

print("List after removing duplicates:", unique_list)
```

```python
#4) Write a program to find elements that are present in the first set but not in the second set.
# Input first set
set1_input = input("Enter elements of the first set separated by space: ")
set1 = set(set1_input.split())

# Input second set
set2_input = input("Enter elements of the second set separated by space: ")
set2 = set(set2_input.split())

# Find difference (elements in set1 but not in set2)
difference_set = set1.difference(set2)

print("Elements in the first set but not in the second set:", difference_set)
```
